```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = 'Car_sales.csv'
data = pd.read_csv(filepath)
data.head()
```

```python
data.info()
```

```python
manufacturer = data['Manufacturer']
sales = data['Sales_in_thousands']
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(15, 10))
plt.barh(manufacturer, sales, color='blue', height=0.8)
plt.xlabel('Car manufaturer')
plt.ylabel('Sale in 1000s')
plt.title('Car sales')
plt.show()
```
