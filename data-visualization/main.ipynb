```python
import pandas as pd  
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('test.csv')
weather
```

```python
weather.head()
```

```python
weather.info() # to see the type of the data
```

```python
sns.barplot(weather['humidity'], weather['temperature'])
```

```python
sns.distplot(weather['humidity']) # used for single distribution
```

```python
sns.jointplot(weather['humidity'], weather['temperature']) # used for more than 2 columns
```

```python
sns.jointplot(weather['humidity'], weather['temperature'], kind='hex')
```

```python
sns.jointplot(weather['humidity'], weather['temperature'], kind='kde')
```

```python
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

```python
sns.stripplot(weather['weather_type'], weather['temperature']) # categorical data
```

```python
sns.stripplot(weather['weather_type'], weather['temperature'], jitter= True) # categorical data
```

```python
sns.boxplot(weather['humidity'], weather['temperature'], hue=weather['weather_type'])
```

```python
sns.barplot(weather['humidity'], weather['temperature'], hue=weather['weather_type'])
```

```python
sns.countplot(weather['weather_type'])
```

```python
sns.pointplot(weather['humidity'], weather['temperature'], hue=weather['weather_type']) # making relationship between more than one column
```

```python
sns.lmplot(x='humidity', y='temperature',data=weather) # it used for linear regression to plot the graph between two variables
```

```python
sns.lmplot(x='humidity', y='temperature', hue='weather_type',data=weather) # it used for linear regression to plot the graph between two variables
```
